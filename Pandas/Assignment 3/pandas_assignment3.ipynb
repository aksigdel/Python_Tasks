Add a new column in countries_df to record the overall GDP per country (product of population & per capita GDP).

```python
df_gdp_countries=df_countries.population*df_countries.gdp_per_capita
df_countries.insert(6,'overall_gdp_per_country',df_gdp_countries,True)
df_countries
```

Create a dataframe containing 10 countries with the lowest GDP per capita, among the counties with population greater than 100 million.


```python
value=100000000
df_10lowest_gdpcountry=(df_countries.query('population > @value').sort_values('gdp_per_capita')).head(10)
df_10lowest_gdpcountry
```

Covid Countries Dataset

```python
df_covid_countries=pd.read_csv('sample_data/covid-countries-data.csv')
df_covid_countries
```

Count the number of countries for which the total_tests data is missing.

```python
df_covid_countries.total_tests.isna().sum()
```

Merge countries_df with covid_data_df on the location column as combined_df.


```python
df_combined=df_countries.merge(df_covid_countries, on='location')
df_combined
```

Add columns tests_per_million, cases_per_million and deaths_per_million into combined_df.


```python
df_combined.insert(9,'tests_per_million', df_combined.total_tests/1000000,True)
```

```python
df_combined.insert(9,'cases_per_million', df_combined.total_cases/1000000,True)
df_combined.insert(9,'deaths_per_million', df_combined.total_deaths/1000000,True)
```

inserting diff col at same loc shifts prev col

```python
df_combined
```

Create a dataframe with 10 countries that have the highest number of tests per million people.


```python
(df_combined.sort_values('tests_per_million',ascending=False).head(10))[['location','tests_per_million']]
```
